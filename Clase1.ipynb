# Chain of Thought
* 3 gráficas de barras (gráficas grises)
* Textos individuales en las barras (reglas %)
* Título
* Subtítulos
* Disclaimer

# Chain of Thought

```python
colors = ["#8c1b5a", "#607cba", "#e18b3b"]
```

```python
fig, axs = plt.subplots(1, 3, figsize=(9,5), sharey=True,)
for n, ax in enumerate(axs):
    sns.barplot(x=36, y=df.index,
                ax=ax, color='lightgray')  # https://matplotlib.org/stable/gallery/color/named_colors.html
    sns.barplot(x=df[df.columns[n]], y=df.index,
                ax=ax, color=colors[n])
    ax.set_title(df.columns[n])
    if n!=0:
      ax.axis('off')
for spine in ["top", "right", "left", "bottom"]:
    axs[0].spines[spine].set_visible(False)
axs[0].tick_params(axis="y", left=False, labelleft=True)
axs[0].tick_params(axis="x", bottom=False, labelbottom=False)
axs[0].set_xlabel("")
axs[0].set_ylabel("")
plt.suptitle("What is the most important issue to you?", ha="right", fontsize=14)
for j in range(3):
 for i in range(df.shape[0]):
    dx = 0
    txt_color = "Black"
    font_w = "normal"
    if int(df.iloc[i, j]) >= 15:
        dx = 6
        txt_color = "White"
        font_w = "bold"
    axs[j].text(df.iloc[i, j] - dx, i, "{}%".format(df.iloc[i, j]), va="center", color=txt_color, weight=font_w)
plt.tight_layout()
plt.show()
```
